<font size = 5>**Import Libraries**</font>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

<font size = 5>**Read Data**</font>

```python
dataset  = pd.read_csv("SampleSuperstore.csv")
dataset.head()
```

```python
dataset.shape
```

<font size = 5>**Missing Values**</font>

```python
dataset.isnull().sum()
```

<font size = 5>**Shipping Mode**</font>

```python
dataset['Ship Mode'].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values

plt.figure(figsize = (6,5))
plt.pie(y, labels = x,autopct = "%0.2f%%")
plt.legend(loc = 1)
plt.title('Shipping Mode')
```

Outcomes ->  
1.Maximum Orders are Shipped by Standard Class  
2.Least Number of Orders are Shipped by Same Day  

<font size = 5>**Types Of Orders**</font>

```python
plt.figure(figsize = (5,3))
sns.countplot(x = "Segment", data = dataset)
plt.title("Types Of Orders")
plt.savefig("fig_name.jpg")
plt.show()
```

Outcomes ->  
1. Maximum Orders are from Consumers  
2. Least Orders are from Home Office  

<font size = 5>**State-wise Sales**</font>

```python
dataset['State'].value_counts()
```

Outcome ->  
1.Top 5 Highest sales States are California, New York, Texas, Pennsylvania, Washington  
2. Five Least Sales State are Wyoming, West Virginia, North Dakota, Maine, District of Columbia   

<font size = 5>**Category-wise Orders**</font>

```python
plt.figure(figsize = (5,3))
sns.countplot(x = "Category", data = dataset)
plt.title("Category-wise Orders")
```

<font size = 5>**Ship Mode As Per Category**</font>

```python
#Sales via Ship mode
plt.figure(figsize = (7,4))
sns.countplot(x = "Ship Mode", data = dataset, hue = "Category")
plt.title("Ship Mode As Per Category")
```

Outcomes ->

2. Highest orders from Furniture are transported from Standard class Ship Mode  
3. Highest orders from Technology are transported from Standard class Ship Mode  
4. Least Orders from Office Supplies are transported from Same Day Ship Mode

6. Least Orders from Technology are transported from Same Day Ship Mode

<font size = 5>**Category wise Sales**</font>

```python
sales_by_category = dataset.groupby('Category')['Sales'].sum()
print(sales_by_category)
```

```python
plt.figure(figsize = (5,3))
sns.barplot(x = "Category",y = "Sales", data = dataset, estimator = 'sum')
plt.title("Category wise Sales")
```

Outcomes ->  
1. Highest Sales is from Technology Category  
2. Least Sales is from Office Supplies Category  

<font size = 5>**Sales by Sub-Category**</font>
